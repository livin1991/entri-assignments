```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_excel('ABCcompany.xlsx')
df
```

```python
df.info
```

```python
df['Height'].head()
```

```python
# Replace 'Height' with random integers between 150 and 180
np.random.seed(42)  # for reproducibility
df['Height'] = np.random.randint(150, 181, size=len(df))

# Verify changes
df['Height'].head()
# Data consistency ensured â€” all height values are now numeric and realistic.
```

```python
team_distribution = df['Team'].value_counts()
team_percentage = round((team_distribution / len(df)) * 100, 2)

team_df = pd.DataFrame({'Team': team_distribution.index,
                        'Count': team_distribution.values,
                        'Percentage': team_percentage.values})
print("Majority of teams have 15 persons, small variation")
team_df
team_df
```

```python
# visualization
plt.figure(figsize=(10,5))
sns.barplot(x='Team', y='Count', data=team_df, palette='viridis')
plt.title('Employee Distribution Across Teams')
plt.xlabel('Team')
plt.ylabel('Number of Employees')
plt.xticks(rotation=90)
plt.show()
print("New Orleans Pelicans have the highest number of employees â€” around 19.\n Employee distribution across teams is relatively balanced,\n with only minor variations â€” suggesting that most teams maintain a similar workforce size, possibly due to standardized operational needs across franchises.")
```

```python
position_distribution = df['Position'].value_counts()

plt.figure(figsize=(10,5))
sns.barplot(x=position_distribution.index, y=position_distribution.values, palette='coolwarm')
plt.title('Employee Segregation by Position')
plt.xlabel('Position')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
bins = [18, 25, 30, 35, 40, 50]
labels = ['18-25', '25-30', '30-35', '35-40', '40-50']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

age_group_distribution = df['Age Group'].value_counts().sort_index()

plt.figure(figsize=(8,5))
sns.barplot(x=age_group_distribution.index, y=age_group_distribution.values, palette='magma')
plt.title('Predominant Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.show()
print("The majority of peoples are youngsters, age between 18-35")

age_group_distribution
```

```python
# By Team
team_salary = df.groupby('Team')['Salary'].sum().sort_values(ascending=False)
# By Position
position_salary = df.groupby('Position')['Salary'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(x=team_salary.index, y=team_salary.values, palette='plasma')
plt.title('Total Salary by Team')
plt.xlabel('Team')
plt.ylabel('Total Salary')
plt.xticks(rotation=90)
plt.show()
print(" higher salary team is Cleveland Cavaliers and lowest salary team is Philadelphia 76ers")


plt.figure(figsize=(10,5))
sns.barplot(x=position_salary.index, y=position_salary.values, palette='crest')
plt.title('Total Salary by Position')
plt.xlabel('Position')
plt.ylabel('Total Salary')
plt.xticks(rotation=0)
plt.show()

team_salary.head(1), position_salary.head(1)
```

```python
corr_value = df[['Age', 'Salary']].corr().iloc[0,1]
print(f"Correlation between Age and Salary = {corr_value:.2f}")

plt.figure(figsize=(6,5))
sns.scatterplot(x='Age', y='Salary', data=df, color='purple')
sns.regplot(x='Age', y='Salary', data=df, scatter=False, color='black')
plt.title('Correlation between Age and Salary')
plt.show()
print(" it is a Positive correlation because As Age increases, there is a slight possibility that Salary also increases.\n but it is also a Weak correlation because The relationship is not very strong, so it is difficult to make reliable predictions.")
```
