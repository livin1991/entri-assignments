```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

titanic = sns.load_dataset("titanic")

sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=titanic, x="embarked", hue="sex")
plt.title("Passenger Count by Embarked Port and Gender")
plt.xlabel("Embarked Port")
plt.ylabel("Passenger Count")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.swarmplot(data=titanic, x="pclass", y="age", hue="sex", dodge=True)
plt.title("Age Distribution across Passenger Classes")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()
```

```python
gender_counts = titanic['sex'].value_counts()
plt.figure(figsize=(5, 5))
plt.pie(
    gender_counts,
    labels=gender_counts.index,
    autopct='%1.1f%%',
    explode=(0.05, 0),
    colors=["#66b3ff", "#ff9999"]
)
plt.title("Gender Distribution")
plt.show()
```

```python
corr = titanic[['age', 'fare', 'pclass', 'sibsp', 'parch', 'survived']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix (Titanic Data)")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.violinplot(data=titanic, x="pclass", y="fare", hue="sex", split=True)
plt.title("Fare Distribution by Class and Gender")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.scatterplot(data=titanic, x="age", y="fare", hue="survived", style="survived")
plt.title("Age vs Fare (Colored by Survival)")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
