```python
"""
AI-Powered Paraphrasing Tool
Author: Livin
Console-based Transformer Paraphraser
Compatible with latest Hugging Face
"""

import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from textblob import TextBlob

# ===============================
# MODEL CONFIGURATION
# ===============================

MODEL_NAME = "t5-small"   # Lightweight & CPU friendly

print("Loading model... Please wait.")

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForSeq2SeqLM.from_pretrained(MODEL_NAME)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

print("Model loaded successfully!\n")


# ===============================
# PARAPHRASING FUNCTION
# ===============================

def paraphrase(text, max_length=128):

    input_text = "paraphrase: " + text

    inputs = tokenizer(
        input_text,
        return_tensors="pt",
        padding=True,
        truncation=True,
        max_length=256
    ).to(device)

    outputs = model.generate(
        **inputs,
        max_length=max_length,
        num_beams=5,
        temperature=1.3,
        top_p=0.9,
        early_stopping=True
    )

    paraphrased_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return paraphrased_text


# ===============================
# GRAMMAR 
# ===============================

def grammar_correct(text):
    blob = TextBlob(text)
    return str(blob.correct())


# ===============================
# MAIN PROGRAM
# ===============================

if __name__ == "__main__":

    print("=== AI Paraphrasing Tool ===\n")

    user_input = input("Enter text to paraphrase:\n")

    print("\nGenerating paraphrase...\n")

    paraphrased = paraphrase(user_input)

    corrected_output = grammar_correct(paraphrased)

    print("\nOriginal Text:")
    print(user_input)

    print("\nParaphrased & Corrected Text:")
    print(corrected_output)
```

```python
"""
Evaluation Module
Includes:
- Semantic Similarity
- BLEU Score
- ROUGE Score
"""

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from nltk.translate.bleu_score import sentence_bleu
from rouge_score import rouge_scorer

print("Loading evaluation model...")

model = SentenceTransformer("all-MiniLM-L6-v2")

print("Evaluation model loaded!\n")


def semantic_similarity(text1, text2):
    emb1 = model.encode([text1])
    emb2 = model.encode([text2])
    score = cosine_similarity(emb1, emb2)[0][0]
    return score


def bleu_score_calc(reference, candidate):
    ref_tokens = reference.split()
    cand_tokens = candidate.split()
    return sentence_bleu([ref_tokens], cand_tokens)


def rouge_score_calc(reference, candidate):
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rougeL'], use_stemmer=True)
    return scorer.score(reference, candidate)


if __name__ == "__main__":

    original = input("Enter original text:\n")
    paraphrased = input("Enter paraphrased text:\n")

    print("\n--- Evaluation Report ---\n")

    similarity = semantic_similarity(original, paraphrased)
    bleu = bleu_score_calc(original, paraphrased)
    rouge = rouge_score_calc(original, paraphrased)

    print(f"Semantic Similarity: {similarity:.4f}")
    print(f"BLEU Score: {bleu:.4f}")
    print("ROUGE Scores:", rouge)
```
